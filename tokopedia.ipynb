```python
from asyncio.windows_events import NULL
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.DataFrame()

page = 1
for iteration in range(0,10000,1):
    
    print(f"page {page}")
    

    boxes = driver.find_elements(By.CLASS_NAME,"css-12sieg3")
    # print(len(boxes))
    for a in boxes:
        descs = a.find_element(By.XPATH,".//div[contains(@class, 'prd_link-product-name css-svipq6')]").text
        pricings = a.find_element(By.XPATH,".//div[contains(@class, 'prd_link-product-price css-1ksb19c')]").text
        # locsold = a.find_element(By.XPATH,"//div[contains(@class, 'css-yaxhi2')]").text
        try:
            loc = a.find_element(By.XPATH,".//span[contains(@class, 'prd_link-shop-loc css-1kdc32b flip')]").text
        except:
            loc = 'NULL'
        
        try:
            shopname = a.find_element(By.XPATH,".//span[contains(@class, 'prd_link-shop-name css-1kdc32b flip')]").text
        except:
            shopname = 'NULL'

        try:
            rating = a.find_element(By.XPATH,".//span[contains(@class, 'prd_rating-average-text css-t70v7i')]").text
        except:
            rating = 'NULL'

        try:
            sold = a.find_element(By.XPATH,".//span[contains(@class, 'prd_label-integrity css-1duhs3e')]").text
        except:
            sold = 'NULL'
        
        
        
        # print(descs, pricings, locsold)
        # df = df.append([descs,pricings,locsold,])
        row = {'description':descs,'pricing':pricings,'locations':loc,'shopname':shopname,'rating':rating,'sold':sold}
        # print(row)
        df = df.append(row,ignore_index=True)
        # df = pd.concat([df, row], ignore_index = True, axis = 0)
        print(len(df))

    element = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Laman berikutnya']")))
    element.location_once_scrolled_into_view
    element.click()
    time.sleep(1)
    page += 1    

    
    
```
